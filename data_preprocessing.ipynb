```python
import pandas as pd
# —Å–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
df = pd.read_csv('/content/drive/MyDrive/Data/posts_ai_telegram.csv', encoding='utf-8')
```

```python
df
```

```python
import re
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from pymorphy3 import MorphAnalyzer
nltk.download('punkt_tab')
nltk.download('stopwords')
pymorphy3_analyzer = MorphAnalyzer()
```

```python

# –æ—á–∏—Å—Ç–∫–∞ –∏ –ø—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
def clean_text(text):
  text = re.sub(r'(?<![–∞-—è—ë–ê-–Ø–Å])-+|-+(?![–∞-—è—ë–ê-–Ø–Å])|[^–∞-—è—ë–ê-–Ø–Å\s-]', '', text)
  text = text.lower().strip()
  return text

df['cleaned_text'] = df['text'].apply(clean_text)

df['cleaned_text'] = df['cleaned_text'].fillna('')

# —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è
df['cleaned_text'] = df['cleaned_text'].apply(word_tokenize)

# –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è
def lemmatize(tokens):
  return [pymorphy3_analyzer.parse(token)[0].normal_form for token in tokens]

df['cleaned_text'] = df['cleaned_text'].apply(lemmatize)

# —Å—Ç–æ–ø-—Å–ª–æ–≤–∞
stops = stopwords.words("russian")
df['cleaned_text'] = df['cleaned_text'].apply(lambda tokens: [token for token in tokens if token not in stops])

df['cleaned_text'] = df['cleaned_text'].str.join(' ')

df
```

```python
#—Å–æ—Ö—Ä–∞–Ω—è–µ–º –Ω–∞—à—É –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö –≤ csv —Ñ–∞–π–ª
df.to_csv('/content/drive/MyDrive/Data/posts_ai_telegram_cleaned.csv', index = False)
```
