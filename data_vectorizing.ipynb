```python
import nltk
import nltk.tokenize
from nltk import word_tokenize
from gensim.models import Word2Vec
nltk.download('punkt_tab')
```

```python
df = df.dropna(subset=['cleaned_text'])
l = df.cleaned_text.to_list()
x_token = [nltk.word_tokenize(text) for text in l]
x_token
```

```python
#–æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ Word2Vec
w2v_model = Word2Vec(sentences=x_token, vector_size=300, window=5, min_count=1, epochs=300)
```

```python
w2v_model.save('/content/drive/MyDrive/W2V_pretrained.bin')
```

```python
# –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç –º–æ–¥–µ–ª–∏ –ø–µ—Ä–µ–¥ –∑–∞–≥—Ä—É–∑–∫–æ–π
w2v_model = Word2Vec()
w2v_model = Word2Vec.load('/content/drive/MyDrive/W2V_pretrained.bin')
```

```python
vocab = w2v_model.wv.key_to_index
vocab_size = len(vocab)
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤ –≤ —Å–ª–æ–≤–∞—Ä–µ: {vocab_size}")
print(vocab)
```

```python
vectors_texts = []
for text in x_token:
    vectors_word = []
    for word in text:
        if word in w2v_model.wv:  #–ø—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ —Å–ª–æ–≤–æ –≤ –º–æ–¥–µ–ª–∏
            vectors_word.append(w2v_model.wv[word])
        else:
            vectors_word.append(np.zeros(w2v_model.vector_size))
    #—É—Å—Ä–µ–¥–Ω—è–µ–º –≤–µ–∫—Ç–æ—Ä–∞
    vector_text = np.mean(vectors_word, axis=0)
    vectors_texts.append(vector_text)

vectors_array = np.array(vectors_texts)
np.save('/content/drive/MyDrive/Data/vectors_w2v.npy', vectors_array)
```
