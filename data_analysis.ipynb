```python
import pandas as pd
# —Å–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
df = pd.read_csv('/content/drive/MyDrive/Data/posts_ai_telegram.csv', encoding='utf-8')
```

```python
df
```

```python
import plotly.express as px

# —Å–æ–∑–¥–∞–µ–º –æ—Ç–¥–µ–ª—å–Ω—É—é –∫–æ–ª–æ–Ω–∫—É —Å —á–∞—Å–æ–º, –¥–æ—Å—Ç–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ date
df['hour'] = pd.to_datetime(df['date']).dt.hour

# —Å–æ–∑–¥–∞–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
fig = px.histogram(df, x='hour', y='views',
                   title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ –ø–æ —á–∞—Å—É –¥–Ω—è', histfunc='sum') # –¥–µ–ª–∞–µ–º —Ç–∞–∫, —á—Ç–æ–±—ã —É –Ω–∞—Å –≤—ã–≤–æ–¥–∏–ª–∞—Å—å –æ–±—â–∞—è —Å—É–º–º–∞ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ –≤—Å–µ—Ö –ø–æ—Å—Ç–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É —á–∞—Å—É

fig.update_layout(
    xaxis_title="–ß–∞—Å—ã",
    yaxis_title="–°—É–º–º–∞ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤",
    bargap=0.1,
    xaxis=dict(tickmode='linear', dtick=1)
)

fig.show()
```

```python
# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç —Ä–µ–∞–∫—Ü–∏–π –∏ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ –æ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤
df['reactions_percent'] = (df['reactions'] / df['views']) * 100
df['comments_percent'] = (df['comments'] / df['views']) * 100


# –°–æ–∑–¥–∞–µ–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É, –ø–µ—Ä–µ–¥–∞–µ–º –≤ x 2 –∑–Ω–∞—á–µ–Ω–∏—è
fig = px.histogram(
    df,
    x=['reactions_percent','comments_percent'],
    title='–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ª–∞–π–∫–æ–≤ –∏ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ –∫ –ø—Ä–æ—Å–º–æ—Ç—Ä–∞–º',
    nbins=20,
    opacity=0.5,
    color_discrete_sequence=['blue', 'green']
)

fig.update_layout(
    yaxis_title="–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å—Ç–æ–≤",
    xaxis_title='% –ª–∞–π–∫–æ–≤ –∏ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤',
    bargap=0.1
)

fig.show()
```

```python
from datetime import date

# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞—Ç—É —Ç–∞–∫, —á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–∏
df['date_str'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')


fig = px.histogram(
    df,
    x='date_str',
    y='views',
    title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º',
    histfunc='sum', # –¥–µ–ª–∞–µ–º —Ç–∞–∫, —á—Ç–æ–±—ã —É –Ω–∞—Å –≤—ã–≤–æ–¥–∏–ª–∞—Å—å –æ–±—â–∞—è —Å—É–º–º–∞ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ –≤—Å–µ—Ö –ø–æ—Å—Ç–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º
    labels={'date_str': '–î–∞—Ç–∞', 'views': '–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤'}
)
fig.update_layout(
    xaxis_title="–î–∞—Ç–∞",
    yaxis_title="–ü—Ä–æ—Å–º–æ—Ç—Ä—ã",
    xaxis=dict(tickangle=45)
)

fig.show()
```
